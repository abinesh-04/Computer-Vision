```python
# ============================
#  Task 4: Optical Character Recognition (OCR)
# ============================

import cv2
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt
import os

# ---- Configure Tesseract Path (Windows only) ----
pytesseract.pytesseract.tesseract_cmd = r"D:\teserract\tesseract.exe"

# ---- Step 1: Ask for image path ----
img_path = input("Enter the image path (e.g., C:/Users/Abinesh/Desktop/text_image.jpg): ").strip().strip('"')

if not os.path.exists(img_path):
    print(" File not found! Check path.")
    exit()

# ---- Step 2: Read and preprocess image ----
image = cv2.imread(img_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# ---- Step 3: OCR using Tesseract ----
text = pytesseract.image_to_string(gray)

# ---- Step 4: Display image and extracted text ----
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Uploaded Image")
plt.show()

print("\n Extracted Text:")
print("---------------------------------------")
print(text)
print("---------------------------------------")
```
